```python
result_df1 = pd.read_csv("2022-04-15.csv")
result_df2 = pd.read_csv("2022-04-15_2.csv")
result_df3 = pd.read_csv("2022-04-15_3.csv")
result_df4 = pd.read_csv("2022-04-15_4_mvoymaguil.csv")
result_df5 = pd.read_csv("2022-04-15_5_cwxccvhugl.csv")
result_df6 = pd.read_csv("2022-04-15_5_mbfbkyocyc.csv")
result_df7 = pd.read_csv("2022-04-15_6_mfzrjvfchj.csv")
result_df8 = pd.read_csv("2022-04-15_7_cpehwbpjfn.csv")
result_df9 = pd.read_csv("2022-04-15_8_rufzqirvbg.csv")
result_df10 = pd.read_csv("2022-04-18_1_kmwaskrksi.csv")
result_df11 = pd.read_csv("2022-04-19_0_ymghvykxfd.csv")
# ignore the 2022-04-19_1   1  1
result_df12 = pd.read_csv("2022-04-19_2_ngtheghypr.csv")
result_df13 = pd.read_csv("2022-04-20_2_flopnwudzj.csv")
result_df = pd.concat([result_df1, result_df2,result_df3,result_df4,result_df5, result_df6, result_df7,result_df8, result_df9,result_df10,result_df11,result_df12,result_df13], ignore_index=True)
result_df["test_rmse"] = result_df.test_mse.apply(np.sqrt)
result_df[["EPOCH","BATCH_SIZE","LR","N_HIDDEN_UNITS","NUM_LAYERS","BI_DIR","data_used_n","min_val_loss_id","min_val_loss","test_mse","test_rmse","note"]].sort_values("test_rmse")

```

```python
best_row = result_df.loc[14]
to_print = f"LSTM, mse {np.round(best_row.test_mse,3)}, rmse {np.round(np.sqrt(best_row.test_mse),3)}"
title = f"LSTM, rmse {np.round(np.sqrt(best_row.test_mse),3)}"
print(title)
# test_rmse = np.array(best_row["rmse"][1:-1].split(", ")).astype(float)
# test_mse = np.array(best_row["test_mse_list"][1:-1].split(", ")).astype(float)
val_loss = np.array(best_row["val_loss_list"][1:-1].split(", ")).astype(float)
train_loss = np.array(best_row["train_loss_list"][1:-1].split(", ")).astype(float)
# x = 

x = np.arange(len(train_loss))
  
# # plot lines
fig = plt.figure(figsize=(10,5)) 
plt.title(title)
plt.plot(x, train_loss, label = "train loss")
plt.plot(x, val_loss, label = "val loss")
plt.legend()
plt.show()
```
